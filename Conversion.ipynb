```python
clean_text1 = []

for word in new_data:
    word = word_tokenize(word)
    clean_text1.append(word)
```

```python
emojis = {':)': 'smile', ':-)': 'smile', ';d': 'wink', ':-E': 'vampire', ':(': 'sad',':-(': 'sad', ':-<': 'sad', ':P': 'raspberry', ':O': 'surprised',
          ':-@': 'shocked', ':@': 'shocked',':-$': 'confused', ':\\': 'annoyed',':#': 'mute', ':X': 'mute', ':^)': 'smile', ':-&': 'confused',
          '$_$': 'greedy','@@': 'eyeroll', ':-!': 'confused', ':-D': 'smile', ':-0': 'yell', 'O.o': 'confused','<(-_-)>': 'robot', 'd[-_-]b': 'dj', 
          ":'-)": 'sadsmile',';)': 'wink',';-)': 'wink', 'O:-)': 'angel','O*-)': 'angel','(:-D': 'gossip', '=^.^=': 'cat'}
```

```python
import re
# Defining regex patterns.
urlPattern        = r"((http://)[^ ]*|(https://)[^ ]*|( www\.)[^ ]*)"
userPattern       = '@[^\s]+'
alphaPattern      = "[^a-zA-Z0-9]"
CLEANR = re.compile('<[^>]+>|[^\w\s]|^\d+\s|\s\d+\s|\s\d+$')

clean_text2 = []

for word in clean_text1:
    clean = []
    for words in word:
        # Replace all URls with 'URL'
        words = re.sub(urlPattern,"",words)
        words = re.sub(CLEANR,"",words)
        # Replace all emojis.
        for emoji in emojis.keys():
            words = words.replace(emoji, "" + emojis[emoji])        
        # Replace @USERNAME to 'USER'.
        words = re.sub(userPattern,'', words)        
        # Replace all non alphabets.
        words = re.sub(alphaPattern, " ", words)
        if words != "":
            clean.append(words)
    clean_text2.append(clean)      
```
