```python
import json
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import wordpunct_tokenize
import pickle
import codecs
```

```python
data = pd.read_excel("Comments_FC.xlsx")
```

```python
new_data = data['Comment']
```

```python
new_data = new_data.astype('str')
```

```python
new_data.head(6)
```

```python
def parseCommentSet(comment_data_path):
    comment_text = []
    english_stopwords_set = set(stopwords.words('english'))
    for word in new_data:
        tokens = wordpunct_tokenize(word)
        words = [word.lower() for word in tokens]
        words_set = set(words)
        comment_elements = words_set.intersection(english_stopwords_set)
        if (len(comment_elements)>1):
            comment_text.append(word)
    
    comment_text_set = set(comment_text)
        
    print(len(comment_text))
    print(len(comment_text_set))
    print(comment_text_set)
    return list(comment_text_set)
```

```python
def translateHinglishText(comment_text):
    counter = 0
    comment_text_translated = []
    n = len(comment_text)
    
    open_file = open("dictionary.pickle", "rb")
    dictionary = pickle.load(open_file)
    open_file.close()

    english_stopwords_set = set(stopwords.words('english'))
    
    for i in range(n):
        text = comment_text[i]
        translated_text = ""
        tokens = wordpunct_tokenize(text)
        words = [word.lower() for word in tokens]
        for word in words:
            if word in english_stopwords_set:
                translated_text = translated_text + " " + word
            elif (word in dictionary):
                #print word + "-" + dictionary[word]
                translated_text = translated_text + " " + dictionary[word]
                counter = counter + 1
            else:
                translated_text = translated_text + " " + word
        comment_text_translated.append(translated_text)
    
    return comment_text_translated
```

```python
def writeToFile(comment_text, comment_text_translated, filename):
    f = codecs.open(filename, 'w', encoding='utf8', errors='ignore')
    n = len(comment_text)
    for i in range(n):
        #f.write(comment_text[i] + '\n')
        f.write(comment_text_translated[i] + '\n')
        f.write('\n')
    f.close()
```

```python
if __name__ == '__main__':
    comment_data_path = new_data
    comment_text = parseCommentSet(comment_data_path)
    comment_text_translated = translateHinglishText(comment_text)
    writeToFile(comment_text, comment_text_translated, 'output.txt')
    
```
